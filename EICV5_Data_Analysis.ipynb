### Creation of  basic tables

```python
#One way tables when the variable is categorical
pd.crosstab(index=df['Gender'], columns ='Frequence')
```

```python
v=pd.crosstab(index=df['Gender'], columns ='Frequence')
(v/v.sum()*100).round(2)
```

# Percentage of male and femaleby provance

```python
#two way tables
pd.crosstab(df['Gender'],df['province'], margins=True, margins_name = 'Total')
```

```python
#In term of percentage
tab = pd.crosstab(df['Gender'],df['province'],margins=True, margins_name = 'Total')
(tab/tab.T['Total']*100).round(2)
```

```python
#In term of percentage
tab = pd.crosstab(df['province'],df['Gender'],margins=True, margins_name = 'Total')
(tab.div(tab['Total'], axis = 0)*100).round(2)
```

# Three way tables
Poverty Status by Gender and Province

```python
pd.crosstab([df.province,df.Gender],df.poverty, margins=True, margins_name = 'Total')
```

```python
#In terms of percentage
tab = pd.crosstab([df.province,df.Gender],df.poverty, margins=True, margins_name = 'Total')
(tab/tab.T['Total']*100).round(2)  #Across the colums
```

```python
#In terms of percentage Across the Rows
tab = pd.crosstab([df.province,df.Gender],df.poverty, margins=True, margins_name = 'Total')
(tab.div(tab['Total'], axis = 0)*100).round(2) 

```

# Pivote tables

Pivote table: is a the combination of categorical values and quantitative values.

```python
#Average consumption by Gender
pd.pivot_table(data = df, index = ['Gender'], values = ['Consumption'], margins=True,margins_name= 'Total',observed=True)
```

```python
#Average consumption by province, urban-rural and by sex
pd.pivot_table(data = df, index = ['province','Gender'], values = ['Consumption'],aggfunc=[np.mean], margins=True,margins_name= 'Total',observed=True)
```

```python
pd.pivot_table(data = df, index = ['Gender'],columns= ['province'], values = ['Consumption'],aggfunc=[np.average,np.std], margins=True,margins_name= 'Total',observed=True)
```
