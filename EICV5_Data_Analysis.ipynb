### Creation of  basic tables

```python
#One way tables when the variable is categorical
pd.crosstab(index=df['Gender'], columns ='Frequence')
```

```python
v=pd.crosstab(index=df['Gender'], columns ='Frequence')
(v/v.sum()*100).round(2)
```

# Percentage of male and femaleby provance

```python
#two way tables
pd.crosstab(df['Gender'],df['province'], margins=True, margins_name = 'Total')
```

```python
#In term of percentage
tab = pd.crosstab(df['Gender'],df['province'],margins=True, margins_name = 'Total')
(tab/tab.T['Total']*100).round(2)
```

```python
#In term of percentage
tab = pd.crosstab(df['province'],df['Gender'],margins=True, margins_name = 'Total')
(tab.div(tab['Total'], axis = 0)*100).round(2)
```

# Three way tables
Poverty Status by Gender and Province

```python
pd.crosstab([df.province,df.Gender],df.poverty, margins=True, margins_name = 'Total')
```

```python
#In terms of percentage
tab = pd.crosstab([df.province,df.Gender],df.poverty, margins=True, margins_name = 'Total')
(tab/tab.T['Total']*100).round(2)  #Across the colums
```

```python
#In terms of percentage Across the Rows
tab = pd.crosstab([df.province,df.Gender],df.poverty, margins=True, margins_name = 'Total')
(tab.div(tab['Total'], axis = 0)*100).round(2) 

```

# Pivote tables

Pivote table: is a the combination of categorical values and quantitative values.

```python
#Average consumption by Gender
pd.pivot_table(data = df, index = ['Gender'], values = ['Consumption'], margins=True,margins_name= 'Total',observed=True)
```

```python
#Average consumption by province, urban-rural and by sex
pd.pivot_table(data = df, index = ['province','Gender'], values = ['Consumption'],aggfunc=[np.mean], margins=True,margins_name= 'Total',observed=True)
```

```python
pd.pivot_table(data = df, index = ['Gender'],columns= ['province'], values = ['Consumption'],aggfunc=[np.average,np.std], margins=True,margins_name= 'Total',observed=True)
```

# Using the above table to create the chart

```python
ax = x.plot(kind='bar', figsize = (10,4),color=['g','y'])
plt.title('Percentage of the population by province', fontweight = "bold")
plt.ylabel("% Percentage")
plt.ylim([0,30]) 
#Adding value labels
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
# using seaborn library

ax=sns.countplot(x ='province', data= df, palette = 'summer', hue='province',
                  order=(['Southern Province', 'Western Province', 'Eastern Province', 'Northern Province', 'Kigali City']))
plt.title('population by province (Counts)')
plt.rcParams['figure.figsize'] = (10,2.5)
plt.xticks(rotation=360)
plt.ylim([0, 19000])

#Adding value labels
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
# The percentage of males and females by province
t = pd.crosstab(df['province'],df['Gender'])
y = round((t/t.sum())*100,2)
y
```

```python
# Bar chart for the above command
ax = y.plot(kind='bar', figsize=(10,3.5), rot=30, color=['b','g'])
plt.title('The percentage of males and females by province')
plt.ylim([0,30]) 
# Value  Labels
for i in ax.containers:
    ax.bar_label(i,fmt='%.1f', fontsize=(12))
plt.show()
```

```python
# We need the total percentages to sum to 100 percent in each province
y['Total'] = y['Male']+y['Female']
y['male_pct'] = (y['Male'] / y['Total']) * 100
y['female_pct'] = (y['Female'] / y['Total']) * 100

# Selecting the columns for males and females
y_pct=y[['male_pct','female_pct']]
# Plotting
ax = y_pct.plot(kind='barh', stacked=False, figsize=(10,4), rot=30, width=0.7,color=['b','#FFA5A5'])
plt.title('The percentage of the males and females by province')
plt.xlim([0,65])
# Adding value Labels
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f', fontsize=(10))
plt.show()
```
